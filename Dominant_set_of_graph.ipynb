```
import networkx as nx
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('ggplot')

def drawAlgoByEdges(G, nodes, edges, chosenEdge, chosenNode):

  #red - chosen node and edges
  #lightgrey - used edges
  #black - unused edges
  #blue - unused nodes
  #light blue - used nodes
  color_map_nodes = [] #colors of each node
  color_map_edges = []  #colors of each edge
  #print("pasirinkta briauna", chosenEdge)
  #print("pasirinkta virsune", chosenNode)
  plt.figure(figsize=(9,9))

  #------- readjusting edges list, so all chosen nodes would appear first in tuple
  j = 0
  for t in edges:
    if t[1] == chosenNode and t[0] not in nodes:
      edges[j] = (t[1], t[0])
    j += 1  

  if chosenEdge[0] != chosenNode:
    temp = chosenEdge
    chosenEdge = (temp[1], temp[0])

  #print(edges)  
  #-------------------------- drawing the graph with highlighted chosen edge and already used nodes, edges ---------

  #for loop for coloring nodes
  counter = 0     #counter is used to know where in color_map_nodes is chosen node
  for node in G:
    if node in nodes and node != chosenNode:  #if node is used, its color is powerblue  
        color_map_nodes.append('powderblue')  

    elif node == chosenNode:    #if node is chosen one, save index for later but color it blue
        color_map_nodes.append('#1f78b4')
        indexOfChosenNode = counter 

    else: 
        color_map_nodes.append('#1f78b4')     #if node is not colored, color it blue
    counter += 1       
  
  # for loop for coloring edges
  for i in G.edges:

    if (i in edges and i[0] != chosenNode)  or ((i[1],i[0]) in edges and i[1] != chosenNode):   # if edge is used and "deleted", color it grey
      color_map_edges.append('lightgrey')

    elif i == chosenEdge or (i[1], i[0]) == chosenEdge:   # if edge is chosen, color it black
      color_map_edges.append('red')

    else:
      color_map_edges.append('black')     #if edge is not used or used but not yet "deleted"


  plt.subplot(221)
  plt.text(0, 1.2, "Pasirenkame briauną:", fontsize=12, ha = "center")      
  nx.draw_shell(G, node_color=color_map_nodes, with_labels=True, edge_color = color_map_edges, font_weight='bold')
  #-----------------------------------------------------------------------------------------------------------------------


  #-------------------------- drawing the graph with highlighted chosen edge, node and already used nodes, edges ---------
  color_map_nodes[indexOfChosenNode] = 'tomato'
  plt.subplot(222)      
  plt.text(0, 1.2, "Pasirenkame galinę viršūnę:", fontsize=12, ha = "center") 
  nx.draw_shell(G, node_color=color_map_nodes, with_labels=True, edge_color = color_map_edges, font_weight='bold')  


  #-------------------------- drawing the graph with highlighted chosen node and edges to be deleted
  j = 0
  for i in G.edges:
    if i in edges:
      if i[0] == chosenNode and color_map_edges[j] != 'lightgrey':        #if edge belongs to chosenNode but wasn't chosen before, color it red, otherwise - grey        
        color_map_edges[j] = 'red'
      else:
        color_map_edges[j] = 'lightgrey'

    elif (i[1],i[0]) in edges:   #if edge belongs to chosenNode but wasn't chosen before, color it red, otherwise - grey       
      if i[1] == chosenNode:
        color_map_edges[j] = 'red'
      else:
        color_map_edges[j] = 'lightgrey' 
    else:
      color_map_edges[j] = 'black'  # if edge wasn't chosen, color it black
    j += 1  
  plt.subplot(223)   
  plt.text(0, 1.4, "Pašaliname visas ", fontsize=12, ha = "center") 
  plt.text(0, 1.2, "incidenčias briaunas:", fontsize=12, ha = "center") 
  nx.draw_shell(G, node_color=color_map_nodes, with_labels=True, edge_color = color_map_edges, font_weight='bold')
  #---------------------------------------------------------------------------------------------------------------------------

#------------------------------------------------------------------drawing the graph after all new changes
  j = 0
  for i in color_map_edges:
   if i == 'red':
      color_map_edges[j] = 'lightgrey'
   j += 1

  color_map_nodes[color_map_nodes.index('tomato')]  = 'powderblue'
  plt.subplot(224)   
  plt.text(-0.5, 1.2, "Naujasis grafas:", fontsize=12, ha = "center")  
  nx.draw_shell(G, node_color=color_map_nodes, with_labels=True, edge_color = color_map_edges, font_weight='bold')
  plt.show()

def drawAlgoByDegrees(G, nodes, chosenNode):

  #red - chosen nodes and edges
  #grey - used edges
  #black - unused edges
  #blue - unused nodes
  #light blue - used nodes
  color_map_nodes = [] #colors of each node
  color_map_edges = []  #colors of each edge
  
  plt.figure(figsize=(9,9))

   #for loop for coloring nodes
  for node in G:
    if node in nodes and node != chosenNode:  #if node is used, its color is powerblue  
        color_map_nodes.append('powderblue')  

    elif node == chosenNode:    #if node is chosen one,  color it red
        color_map_nodes.append('tomato')

    else: 
        color_map_nodes.append('#1f78b4')     #if node is not colored, color it blue    
  

  # for loop for coloring edges
  for i in G.edges:
    if (i[0] in nodes and i[0] != chosenNode) or (i[1] in nodes and i[1] != chosenNode):   # if edge is used, color it grey
      color_map_edges.append('lightgrey')  
    else:
      color_map_edges.append('black')     #if edge is not used, color it black

  plt.subplot(221)
  plt.text(0, 1.4, "Pasirenkame viršūnę su ", fontsize=12, ha = "center") 
  plt.text(0, 1.2, "aukščiausiu laipsniu:", fontsize=12, ha = "center")      
  nx.draw_shell(G, node_color=color_map_nodes, with_labels=True, edge_color = color_map_edges, font_weight='bold')

  j = 0
  for i in G.edges:      #all edges connected by node color red
    if (i[0] == chosenNode and i[1] not in nodes) or (i[1] == chosenNode and i[0] not in nodes):
      color_map_edges[j] = "red"
    j += 1  

  plt.subplot(222)
  plt.text(0, 1.4, "Pašaliname visas ", fontsize=12, ha = "center") 
  plt.text(0, 1.2, "incidenčias briaunas:", fontsize=12, ha = "center")       
  nx.draw_shell(G, node_color=color_map_nodes, with_labels=True, edge_color = color_map_edges, font_weight='bold') 

  j = 0
  for i in color_map_edges:    #"delete" edges and color them grey, nodes - light blue
    if i == "red":
      color_map_edges[j] = "lightgrey"
    j += 1
  color_map_nodes[color_map_nodes.index('tomato')]  = 'powderblue'        

  plt.subplot(223)   
  plt.text(-0.5, 1.2, "Naujasis grafas:", fontsize=12, ha = "center")  
  nx.draw_shell(G, node_color=color_map_nodes, with_labels=True, edge_color = color_map_edges, font_weight='bold')
  plt.show()

def edgeMatrix(G):
  print("Sukuriame gretimumo matrica:")
  numberOfNodes = nx.number_of_nodes(G)
  listOfEdges = list(G.edges) #grafo G briaunu sarasas
  print("Briaunu sarasas: ", listOfEdges)
  A = np.matrix(np.zeros((numberOfNodes,numberOfNodes), dtype='double')) #"tuscia " matrica
  for i in listOfEdges: # uzpildome ja duomenimis
    x = i[0] - 1
    y = i[1] - 1
    A[x,y] = 1
    A[y,x] = 1
  print("Gretimumo matrica: ")
  print(A)  
  return A  

def highestDegree(A):
  oldNum = 0
  v = -1
  size = len(A)
  for i in range(size):
    newNum = 0
    for j in range(size):  
      if A[i,j] == 1:
        newNum += 1                     #skaiciuojame briaunu skaiciu
    if newNum >= oldNum and newNum != 0:        #jei didesnis nei ankstesnes virsune ir nelygus nuliui, rezultata issisaugome 
      oldNum = newNum  
      v = i
  #print(v, oldNum)    
  if v == -1:                        #jei -1 vadinasi nebera briaunu
    return -1   
  print("Ieskome virsunes su auksciausiu laipsniu: virsune = ",  v + 1, ", jos laipsnis = ", oldNum)  
  return v + 1 
        

def dominantSetOfGraphByDegrees(G):
  A = edgeMatrix(G)
  B = []
  print("Dominuojanciuju virsuniu aibe: DS = {", B , "}")
  numberOfNodes = nx.number_of_nodes(G)
  for i in range(numberOfNodes):
    deg = highestDegree(A)
    #print(deg)
    if deg != -1:
      B.append(deg)
      drawAlgoByDegrees(G, B, deg)
      print("Dominuojanciuju virsuniu aibe: DS = {", B , "}")
      for j in range(numberOfNodes):
        A[B[i] - 1, j] = 0
        A[j, B[i]-1] = 0
      #print(A)
  return B          

def dominantSetOfGraphByEdges(G):
  A = edgeMatrix(G)
  B = []
  listOfEdges = list(G.edges) #grafo G briaunu sarasa
  listOfUsedEdges = []
  nx.draw_shell(G, with_labels=True, font_weight='bold')
  plt.show()
  print("Dominuojanciuju virsuniu aibe: DS = {", B , "}")
  numberOfNodes = nx.number_of_nodes(G)
  for i in listOfEdges:
    if i[0] not in B or i[1] not in B:        #patikriname ar bent virsune nebuvo paimta
  
      if i[0] not in B:
          x = i[0]

      elif i[1] not in B:
          x = i[1] 
   
      check = 0
      for j in range(numberOfNodes):                               
            if A[x - 1, j] == -1 or A[j, x - 1] == -1:      #tikriname ar virsune neturi pasalintu briaunu (kad galimai rastu mazesne aibe)
              check = -1
              break
      if check == 0:                                    #jei virsune neturi pasalintu briaunu, itraukiame ja i aibe 
        chosenEdge = i 
        chosenNode = x     
        B.append(x)
        listOfUsedEdges.append(i)
        print("Pasirinkta briauna : " , i, " ir virsune:" , x, " --->Dominuojanciuju virsuniu aibe: DS = {", B , "}")
        
        for j in range(numberOfNodes):
          if A[x - 1, j] == 1:                                               #pazymime pasalintas briaunas
            A[x - 1, j] = -1
            A[j, x - 1] = -1
            if (x, j+1) not in listOfUsedEdges and (j+1, x) not in listOfUsedEdges:
              listOfUsedEdges.append((x,j+1))
        drawAlgoByEdges(G, B, listOfUsedEdges, chosenEdge, chosenNode)
  for i in range(numberOfNodes):                               # praeiname likusias briaunas ir pazymime pasalintas
     if i + 1 not in B:
       found = False
       for j in range (numberOfNodes):
         if A[i,j] == 1:
           A[i, j] = -1
           A[j, i] = -1
           if found == False:
              chosenEdge = (i+1, j+1)
              found = True
           listOfUsedEdges.append((i+1,j+1))
       if found == True:
          B.append(i + 1) 
          chosenNode = i + 1
          print("Dominuojanciuju virsuniu aibe: DS = {", B , "}")
          drawAlgoByEdges(G, B, listOfUsedEdges, chosenEdge, chosenNode)
  return B
G = nx.Graph() 
G.add_edges_from([(1,2),(2,3),(3,4),(3,5),(5,6),(5,7),(5,8),(8,9),(2,5)])
B = nx.Graph()
B.add_edges_from([(1,2),(1,7),(1,3),(4,2),(2,8),(5,3),(3,4),(4,6),(5,6),(5,7),(6,8), (7,8)])

print ("DOMINUOJANČIOSIOS VIRŠŪNIŲ AIBĖS PAIEŠKOS ALGORITMAS PASIRENKANT AUKŠČIAUSIĄ VIRŠŪNIŲ LAIPSNĮ\n\n")
print("Galutine aibe: DS = {", dominantSetOfGraphByDegrees(G), "}\n\n")
print("Galutine aibe: DS = {", dominantSetOfGraphByDegrees(B), "}\n\n")

print ("\n\n\n DOMINUOJANČIOSIOS VIRŠŪNIŲ AIBĖS PAIEŠKOS ALGORITMAS PASIRENKANT BRIAUNĄ-----------------------------------------------------------------------\n\n")

print("Galutine aibe: DS = {", dominantSetOfGraphByEdges(G), "}\n\n")
print("Galutine aibe: DS = {", dominantSetOfGraphByEdges(B), "}\n\n")
```
